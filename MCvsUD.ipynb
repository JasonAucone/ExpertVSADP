```python
import pandas as pd

# data imports
mcqb = pd.read_excel("/Users/jason/Documents/Project/MCQB.xlsx")
mcrb = pd.read_excel("/Users/jason/Documents/Project/MCRB.xlsx")
mcwr = pd.read_excel("/Users/jason/Documents/Project/MCWR.xlsx")
mcte = pd.read_excel("/Users/jason/Documents/Project/MCTE.xlsx")
ud = pd.read_excel("/Users/jason/Documents/Project/UDADP.xlsx")

# checks to make sure the import works
print(mcqb.head())
print(mcrb.head())
print(mcwr.head())
print(mcte.head())
print(ud.head())
```

```python
# merge the underdog data with mike clay's data
qb_matched = pd.merge(mcqb, ud, left_on='Quarterback', right_on='Player', how='inner')

# cut off the qb for plotting reasons
qb_matched['positionRank'] = qb_matched['positionRank'].str.replace('QB', '', regex=False).astype(int)

print(qb_matched.head())
```

```python
# merge the underdog data with mike clay's data
rb_matched = pd.merge(mcrb, ud, left_on='RunningBack', right_on='Player', how='inner')

# cut off the rb for plotting reasons
rb_matched['positionRank'] = rb_matched['positionRank'].str.replace('RB', '', regex=False).astype(int)

print(rb_matched.head())
```

```python
# merge the underdog data with mike clay's data
wr_matched = pd.merge(mcwr, ud, left_on='WideReceiver', right_on='Player', how='inner')

# cut off the wr for plotting reasons
wr_matched['positionRank'] = wr_matched['positionRank'].str.replace('WR', '', regex=False).astype(int)

print(wr_matched.head())
```

```python
# merge the underdog data with mike clay's data
te_matched = pd.merge(mcte, ud, left_on='TightEnd', right_on='Player', how='inner')

# cut off the te for plotting reasons
te_matched['positionRank'] = te_matched['positionRank'].str.replace('TE', '', regex=False).astype(int)

print(te_matched.head())
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# set the style
sns.set(style="whitegrid")

# create the plot
plt.figure(figsize=(12, 8))

# spread the text out using a horizontal jitter
for _, row in qb_matched.iterrows():
    x_jitter = row['positionRank'] + np.random.uniform(-0.3, 0.3)  # Adjust range as needed
    plt.text(x_jitter, row['PosRk'], row['Quarterback'], fontsize=8, ha='center')

# add 1:1 reference line
lims = [min(qb_matched['positionRank'].min(), qb_matched['PosRk'].min()),
        max(qb_matched['positionRank'].max(), qb_matched['PosRk'].max())]
plt.plot(lims, lims, 'k--', alpha=0.7, zorder=0)

# labels and titles
plt.xlabel("Underdog's Rankings", fontsize=12)
plt.ylabel("Mike's Rankings", fontsize=12)
plt.title('Mike Clay VS Underdog ADP (QBs)', fontsize=16)

plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(12, 8))

for _, row in rb_matched.iterrows():
    x_jitter = row['positionRank'] + np.random.uniform(-0.3, 0.3)  # Adjust range as needed
    plt.text(x_jitter, row['PosRk'], row['RunningBack'], fontsize=8, ha='center')

lims = [min(rb_matched['positionRank'].min(), rb_matched['PosRk'].min()),
        max(rb_matched['positionRank'].max(), rb_matched['PosRk'].max())]
plt.plot(lims, lims, 'k--', alpha=0.7, zorder=0)

plt.xlabel("Underdog's Rankings", fontsize=12)
plt.ylabel("Mike's Rankings", fontsize=12)
plt.title('Mike Clay VS Underdog ADP (RBs)', fontsize=16)

plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(12, 8))

for _, row in wr_matched.iterrows():
    x_jitter = row['positionRank'] + np.random.uniform(-0.3, 0.3)  # Adjust range as needed
    plt.text(x_jitter, row['PosRk'], row['WideReceiver'], fontsize=8, ha='center')

lims = [min(wr_matched['positionRank'].min(), wr_matched['PosRk'].min()),
        max(wr_matched['positionRank'].max(), wr_matched['PosRk'].max())]
plt.plot(lims, lims, 'k--', alpha=0.7, zorder=0)

plt.xlabel("Underdog's Rankings", fontsize=12)
plt.ylabel("Mike's Rankings", fontsize=12)
plt.title('Mike Clay VS Underdog ADP (WRs)', fontsize=16)

plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(12, 8))

for _, row in te_matched.iterrows():
    x_jitter = row['positionRank'] + np.random.uniform(-0.3, 0.3)  # Adjust range as needed
    plt.text(x_jitter, row['PosRk'], row['TightEnd'], fontsize=8, ha='center')

lims = [min(te_matched['positionRank'].min(), te_matched['PosRk'].min()),
        max(te_matched['positionRank'].max(), te_matched['PosRk'].max())]
plt.plot(lims, lims, 'k--', alpha=0.7, zorder=0)

plt.xlabel("Underdog's Rankings", fontsize=12)
plt.ylabel("Mike's Rankings", fontsize=12)
plt.title('Mike Clay VS Underdog ADP (TEs)', fontsize=16)

plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Calculate the difference
qb_matched['rank_diff'] = qb_matched['PosRk'] - qb_matched['positionRank']

# Sort by the difference descending
qb_sorted = qb_matched.sort_values(by='rank_diff', ascending=False)

# Print the sorted result
print(qb_sorted[['Quarterback', 'PosRk', 'positionRank', 'rank_diff']])
```
